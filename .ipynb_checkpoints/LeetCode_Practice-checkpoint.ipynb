```python
# 303. Range Sum Query - Immutable

class NumArray:
    def __init__(self, nums):
        print("\n==============================")
        print("Initializing NumArray with:", nums)

        # Calculate prefix sums
        self.prefix_sums = [0] * (len(nums) + 1)

        for i in range(len(nums)):
            self.prefix_sums[i + 1] = self.prefix_sums[i] + nums[i]
            print(f"Step {i}: prefix_sums[{i+1}] = {self.prefix_sums[i]} + {nums[i]} = {self.prefix_sums[i+1]}")

        print("Final prefix_sums:", self.prefix_sums)

    def sumRange(self, i, j):
        result = self.prefix_sums[j + 1] - self.prefix_sums[i]
        print(f"\nsumRange({i}, {j}) -> prefix[{j+1}] - prefix[{i}] = {self.prefix_sums[j+1]} - {self.prefix_sums[i]} = {result}")
        return result


# ==============================
# TEST CASES
# ==============================

print("\nTEST CASE 1")
nums = [-2, 0, 3, -5, 2, -1]
obj = NumArray(nums)
obj.sumRange(0, 2)
obj.sumRange(2, 5)
obj.sumRange(0, 5)


print("\nTEST CASE 2")
nums = [1, 2, 3, 4, 5]
obj = NumArray(nums)
obj.sumRange(1, 3)


print("\nTEST CASE 3")
nums = [-1, -2, -3, -4]
obj = NumArray(nums)
obj.sumRange(0, 3)


print("\nTEST CASE 4")
nums = [10]
obj = NumArray(nums)
obj.sumRange(0, 0)


print("\nTEST CASE 5")
nums = [0, 0, 0, 0]
obj = NumArray(nums)
obj.sumRange(0, 3)


print("\nTEST CASE 6")
nums = [100, -50, 200, -150, 300]
obj = NumArray(nums)
obj.sumRange(1, 4)


print("\nTEST CASE 7")
nums = [1, -1, 1, -1, 1, -1]
obj = NumArray(nums)
obj.sumRange(0, 5)


print("\nTEST CASE 8")
nums = [5, 10, 15, 20]
obj = NumArray(nums)
obj.sumRange(0, 3)
obj.sumRange(0, 3)


print("\nTEST CASE 9")
nums = list(range(1, 11))
obj = NumArray(nums)
obj.sumRange(3, 8)


print("\nTEST CASE 10")
nums = [3, -2, 5, -1, 6, -3]
obj = NumArray(nums)
obj.sumRange(2, 4)
obj.sumRange(1, 3)
```
